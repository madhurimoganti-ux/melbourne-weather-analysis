```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5,4))

sns.heatmap(data.corr(), annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap - Melbourne Weather")
plt.show()
```

## Final Insight

- Temperature and rainfall have a weak negative correlation in Melbourne
- Weather patterns are not strongly dependent on temperature alone
- Multiple atmospheric factors influence rainfall

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=data,
    x="MaxTemp",
    y="Rainfall",
    alpha=0.4
)

plt.title("Temperature vs Rainfall in Melbourne")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.regplot(
    data=data,
    x="MaxTemp",
    y="Rainfall",
    scatter_kws={"alpha":0.3},
    line_kws={"color":"red"}
)

plt.title("Temperature vs Rainfall with Trend Line")
plt.show()
```
